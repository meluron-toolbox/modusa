# Signal Operations

```python
import modusa as ms
```

```python
import numpy as np
```

```python
signal = ms.generator1D.randint(0, 10, 20, sr=10)
signal_ones = ms.generator1D.ones_like(signal)
```

```python
display(signal.plot(fmt="bo:", legend="Signal1", title=""))
display(signal_ones.plot(fmt="ko:", legend="Signal2", title=""))
display((signal * signal_ones).plot(fmt="ro:", legend="$Signal1 * Signal2$", title=""))
display((signal + signal_ones).plot(fmt="ro:", legend="$Signal1 + Signal2$", title=""))
```

```python
from modusa.io import AudioLoader
from modusa.generators import AudioWaveformGenerator
```

```python
signal1 = AudioWaveformGenerator.generate_sinusoid(f=10, duration=4, sr=10000)
signal2 = AudioWaveformGenerator.generate_sinusoid(f=20, duration=4, sr=10000)
signal = signal1 | signal2 | signal1 + signal2 | signal1 * signal2
signal.title = "Mixed Signals"
```

```python
signal.plot(highlight_regions=[(3.5, 4.5, "Transition")], legend=("Mixed signals", "upper right"))
```

```python
spec = signal.to_spectrogram(n_fft=1024, hop_length=512).magnitude
spec
```

```python
spec.plot(highlight_regions=[(3.5, 4.5, "Transition")], ylim=(0, 100), Mlabel="Magnitude Coef")
```

```python
signal.play(regions=[(0, 5), (6, 7)])
```
