### Working with Audio Signal

```python
from modusa.signals import AudioSignal
import numpy as np
```

```python
signal = AudioSignal.from_array_with_sr(np.random.random(100), 2)
signal.set_units(t_unit="sec")
```

```python
signal.plot(linestyle="--", color="k", highlight=[(20, 40)])
```

```python
signal[int(20*signal.sr):int(40*signal.sr):1].plot(linestyle=":")
```

```python
signal.sr
```

```python
signal.t
```

```python
signal[1:10:1.5].plot(marker="o")
```

```python
signal[1:10:1].plot(marker="o")signal[1:10:2].plot(marker="o")
```
