```python
import numpy as np
import modusa as ms
```

```python
sr = 1000
t = ms.sax.linear(5*sr, sr, label="Time (sec)")
```

```python
s = (2 * np.pi * 10 * t).sin()
```

```python
display(s.plot())
display(s.floor().plot())
display((s / s.mean()).plot())
display((s - s).plot())
display((s.cos()**2 + s.sin()**2).plot())
```

```python
(s / s.max()).plot()
```
